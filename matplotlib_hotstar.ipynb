```python
# chart - 1
# List the Tv series by Age Rating
plt.figure(figsize=(16,8))
Tv['Age_rating'].unique()
label=['U/A 13+', 'U/A 7+', 'U/A 16+', 'U', 'A', 'PG']
Tv1=Tv['Title'].groupby(Tv['Age_rating']).size()
plt.pie(Tv1,autopct='%.2f',labels=label)
plt.title('Distribution by Age_rating')
plt.show()
```

```python
year_genre=hot.groupby(['Year','Type']).size().sort_index().reset_index()
year_genre=year_genre.rename(columns={0:'count'})
year_genre.head()
```

```python
year_genre=hot.groupby(['Year','Type']).size().sort_index().reset_index()
year_genre=year_genre.rename(columns={0:'count'})
year_genre.head()
```

```python
# chart - 15
# No. of shows telecasted from year to year
min_range=[min(hot.Year)-1,1950,1975,2000,max(hot.Year)]
max_range=[1950,1975,2000,2023]
fig,axes=plt.subplots(1,4, figsize=(30,8))
for min_year,max_year, axs in zip(min_range,max_range, axes):
    df=year_genre.loc[(year_genre.Year>min_year) & (year_genre.Year<=max_year) ,:]
    g=df.pivot(index='Year',columns='Type', values='count').plot(kind='line', ax=axs)
    g.set_title(f'from year {min_year} to year {max_year}')
    g.set_ylabel('Number of Movie & TV')
    g.legend(('Movies','Tv'))
```
